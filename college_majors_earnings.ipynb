# Recent College Graduates Analysis

This analysis will look at data regarding recent college graduates such as major, median salary, number of men, number of women, full and part time employment, and more.

```python
import pandas as pd
import matplotlib as plt

# displays plots in line
%matplotlib inline

# read the csv file into a pandas dataframe
recent_grads = pd.read_csv('recent-grads.csv')
```

```python
# explore the first and last few rows using df.head() and df.tail()
recent_grads.head()
```

```python
# create quick summary statistics for the data
recent_grads.describe()
```

```python
# return the number of rows in the data set
raw_count = len(recent_grads)
print(raw_count)
```

```python
# drop rows with NULL values
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
print(cleaned_data_count)
```

```python
# create 6 scatter plots exploring the relationship between various columns
ax1 = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax1.set_title('Sample Size and Median')

ax2 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax2.set_title('Sample Size and Unemployment Rate')

ax3 = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax3.set_title('Full Time and Median')

ax4 = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax4.set_title('Share Women and Unemployment Rate')

ax5 = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax5.set_title('Men and Median')

ax6 = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax6.set_title('Women and Median')
```

# Scatterplot analysis

The following points can be noted after analysis from the scatter plot data:
 - There is no relationship between major popularity and median salary
 - There is no relationship between majors that were majority female and median salary
 - There is no relationship between the number of full time employees and median salary

# Histogram analysis

Base on the histograms above, there are 2 important points:

 - Approximately 55% of majors are predominantly female, and approximately 45% predominantly male. There are 3 majors that are only male and 8 majors that are only female.
