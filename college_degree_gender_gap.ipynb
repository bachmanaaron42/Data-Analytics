# College Degree gender gap analysis

This analysis will visualize the percentage of men vs. the percentage of women that receive degrees from various categories.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# read csv file into data frame
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

# initialize color blind colors
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
h_line_color = (171/255, 171/255, 171/255)

# differentiate degrees for each column of figure grid
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

#initialize figure and size
fig = plt.figure(figsize=(9, 18))

# loop through the left most column of the figure grid, and plot the axes

incr_one = 0

for sp in range(1, 17, 3):
    ax = fig.add_subplot(6,3,sp)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[incr_one]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[incr_one]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[incr_one])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.set_yticks([0, 100])
    ax.axhline(50, c=h_line_color, alpha=0.3)
    
    if sp == 1:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 16:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')
        
    incr_one += 1

# loop through the middle column of the figure grid, and plot the axes
    
incr_two = 0    
    
for num in range(2, 15, 3):
    ax = fig.add_subplot(6, 3, num)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[incr_two]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[incr_two]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(lib_arts_cats[incr_two])
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    ax.tick_params(bottom='off', top='off', left='off', right='off', labelbottom='off')
    ax.set_yticks([0, 100])
    ax.axhline(50, c=h_line_color, alpha=0.3)
    
    if num == 2:
        ax.text(2000, 80, 'Women')
        ax.text(2000, 20, 'Men')
    elif num == 14:
        ax.tick_params(labelbottom='on')
        
    incr_two += 1

# loop through the right most column of the figure grid, and plot the axes
    
incr_three = 0
    
for i in range(3, 19, 3):
    ax = fig.add_subplot(6, 3, i)
    ax.plot()
    ax.plot(women_degrees['Year'], women_degrees[other_cats[incr_three]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[incr_three]], c=cb_orange, label='Men', linewidth=3)
    ax.set_title(other_cats[incr_three])
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    ax.tick_params(bottom='off', top='off', left='off', right='off', labelbottom='off')
    ax.set_yticks([0, 100])
    ax.axhline(50, c=h_line_color, alpha=0.3)
    
    if i == 3:
        ax.text(2000, 72, 'Women')
        ax.text(2000, 25, 'Men')
    elif i == 18:
        ax.text(2000, 25, 'Women')
        ax.text(2000, 65, 'Men')
        ax.tick_params(labelbottom='on')
    
    incr_three += 1

# save the figure to an image within the same directory we are working in
plt.savefig('gender_degrees.png')
plt.show()
```

# Observations

As you can see from the visualiations above, the percentage of men receiving engineering and computer science degrees are higher than the percentage of women receiving those degrees, whereas the percentage of women receiving health professions and education degrees are higher than the percentage of men receiving those degrees.
